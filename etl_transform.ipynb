## TRANSFORM FULL EXTRACTION DATA


```python
import pandas as pd

# Load full extracted data
df_full = pd.read_csv('custom_data.csv')

# 1. Clean: Drop missing values and duplicates
df_full.dropna(inplace=True)
df_full.drop_duplicates(inplace=True)

# 2. Enrich: Create total_contribution column
df_full['total_contribution'] = (
    df_full['points_scored'] + df_full['assists'] + df_full['rebounds']
)

# 3. Structure: Convert date columns to datetime
df_full['match_date'] = pd.to_datetime(df_full['match_date'])
df_full['last_updated'] = pd.to_datetime(df_full['last_updated'])

# Save transformed data
df_full.to_csv('transformed_full.csv', index=False)

df_full.head()
```

## TRANSFORM INCREMENTAL DATA
